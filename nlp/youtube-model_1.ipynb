## Text Vectorization

```python
tfidf_vectorizer_params = TfidfVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 3), max_df = .5)
```

```python
tfidf_vectorizer = tfidf_vectorizer_params.fit_transform(df_comments)
```

```python
pd.DataFrame(tfidf_vectorizer.toarray(), columns=tfidf_vectorizer_params.get_feature_names())
```

```python
pickle.dump(tfidf_vectorizer, open('tfidf_vectorizer.pickle', 'wb'))
```

```python
vects_cos_sim = cosine_similarity(tfidf_vectorizer, tfidf_vectorizer)
```

```python
pd.DataFrame(data=vects_cos_sim, index=df_comments, columns=df_comments).head()
```

## Recommendation Engine
The engine is expected to carry out 4 fairly simple tasks:

- Get the index of the requested show title.
- Compute the cosine similarity of all shows to this show.
- Sort the results from step 2 from highest cosine similarity to lowest and get the top 5.
- Return the shows titles and similarity confidence (cosine similarity) in JSON format.

```python
def recommended_shows(comments):
    
    #Get show index
    title_iloc = df.index[df['textdisplay'] == comments][0]
    
    #Get cosine similarity
    show_cos_sim = cosine_similarity(tfidf_vectorizer[title_iloc], tfidf_vectorizer).flatten()
    
    #Get the top 5 most similar shows
    sim_comments_vects = sorted(list(enumerate(show_cos_sim)), key=lambda x: x[1], reverse=True)[1:6]
    
    #Return result
    response = '\n'.join([f'{df.iloc[t_vect[0]][0]} --> confidence: {round(t_vect[1],1)}' for t_vect in sim_comments_vects])
    
    return response
print(recommended_shows('So good'))
```
